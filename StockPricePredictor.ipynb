```
import numpy as np
import matplotlib.pyplot as mpl
from sklearn.preprocessing import scale
from TFANN import ANNR
from google.colab import files
```

```
files.upload()
```

```
stock_data = np.loadtxt('TCS1.csv', delimiter=",", skiprows=1, usecols = (0,4))   
stock_data = scale(stock_data)

prices = stock_data[:,1].reshape(-1,1)
dates = stock_data[:,0].reshape(-1,1)

mpl.plot(dates[:,0] , prices[:,0])
mpl.show()
```

```
input = 1
output =1 
hidden = 60

layers = [('F', hidden), ('AF', 'tanh'), ('F', hidden), ('AF', 'tanh'), ('F', hidden), ('AF', 'tanh'), ('F', output)]
mlpr = ANNR([input], layers, batchSize = 200, maxIter = 20000, tol = 0.2, reg = 1e-4, verbose = True)
```

```

totalDays = len(dates)

mlpr.fit(dates[0:(totalDays)], prices[0:(totalDays)])
```

```
pricePredict = mlpr.predict(dates)

mpl.plot(dates, prices)
mpl.plot(dates, pricePredict, c='#5aa9ab')
mpl.show()
```

```
input2 = 1
output2 = 1
hidden2 = 60
 
layers = [('F', hidden2), ('AF', 'tanh'), ('F', hidden2), ('AF', 'tanh'), ('F', hidden2), ('AF', 'tanh'), ('F', output2)]

mlpr2 = ANNR([input2], layers, batchSize = 200, maxIter = 10000, tol = 0.05, reg = 1e-4, verbose = True)
```

```

totalDays = len(dates)
mlpr2.fit(dates[0:(totalDays)], prices[0:(totalDays)])
```

```
pricePredict2 = mlpr2.predict(dates)
mpl.plot(dates, prices)
mpl.plot(dates, pricePredict, c='#5aa9ab')
mpl.plot(dates, pricePredict2, c='#8B008B')
mpl.show()
```
